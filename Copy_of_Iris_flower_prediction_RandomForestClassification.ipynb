```python
# Memuat dataset Iris menggunakan load_iris() dari scikit-learn.
# Dataset ini berisi informasi tentang tiga spesies Iris: Iris-setosa, Iris-versicolor, dan Iris-virginica.
from sklearn.datasets import load_iris

# Mengimpor RandomForestClassifier dari sklearn.ensemble
# Untuk menginisialisasi model Random Forest yang akan digunakan untuk melakukan klasifikasi pada dataset.
from sklearn.ensemble import RandomForestClassifier

# Mengimpor pandas untuk memanipulasi data dalam bentuk DataFrame. dengan alias pd
import pandas as pd

# Mengimpor numpy dengan alias np.
import numpy as np

# Menetapkan seed acak dengan np.random.seed(0) untuk hasil yang dapat direproduksi.
# Berguna untuk memastikan hasil yang sama setiap kali kode dijalankan.
np.random.seed(0)
```

```python
# Membuat objek data_iris yang berisi data Iris.
data_iris = load_iris()

# Membuat DataFrame data_frame yang berisi data fitur data_iris, dengan nama kolom yang diambil dari data_iris.feature_names.
data_frame = pd.DataFrame(data_iris.data, columns = data_iris.feature_names)

# Menampilkan 5 baris teratas.
data_frame.head()
```

```python
# Menambahkan kolom 'spesies' ke DataFrame data_frame untuk menampilkan nama spesies berdasarkan data target dari dataset data_iris.
data_frame['spesies'] = pd.Categorical.from_codes(data_iris.target, data_iris.target_names)

# Menampilkan 5 baris teratas.
data_frame.head()
```

```python
# Membuat kolom 'is_train' yang berisi bilangan acak antara 0 dan 1. hal ini dilakukan untuk membagi data menjadi data pelatihan dan data pengujian.
# Sekitar 75% data pada kolom is_train akan dianggap sebagai data pelatihan dan sebaliknya. sekitar 25% sisanya dianggap sebagai data pengujian.
data_frame['is train'] = np.random.uniform(0, 1, len(data_frame)) <= 0.75

# Menampilkan 5 baris teratas.
data_frame.head()
```

```python
print(int(150*(0.75)))
```

```python
# Membuat DataFrames terpisah untuk data pelatihan (latih) dan data pengujian (uji).
latih = data_frame[data_frame['is train']==True]
uji = data_frame[data_frame['is train']==False]

# Menampilkan jumlah data pelatihan dan data pengujian.
print(f'Jumlah data pelatian sebanyak: {len(latih)} data')
print(f'Jumlah data pengujian sebanyak: {len(uji)} data')
```

```python
# Menentukan kolom fitur yang akan digunakan dalam pelatihan model, yaitu kolom-kolom yang ada dalam data_frame kecuali kolom 'species'.
# karena kolom species berada pada kolom ke 5, maka diambil kolom 1 sampai 4
fitur = data_frame.columns[:4]

# Menampilkan kolom fitur
print(fitur)
```

```python
# Mengubah nama spesies menjadi angka menggunakan pd.factorize() dan menyimpannya dalam variabel y
x = pd.factorize(latih['spesies'])[0]

# Menampilkan data spesies yang sudah diubah menjadi angka.
# Dengan setosa(0), versicolor(1), dan virginica(2).
# print(x)

# Membuat DataFrame baru untuk menampilkan hasil faktorisasi
faktorisasi_df = pd.DataFrame({'spesies_angka': x})

# Menampilkan DataFrame hasil faktorisasi
print(faktorisasi_df)
```

```python
# Membuat objek classifier dari RandomForestClassifier dengan beberapa parameter, seperti jumlah pekerjaan (n_jobs) dan seed acak.
klasifikasi = RandomForestClassifier(n_jobs=2, random_state=0)

# Melatih model dengan menggunakan data pelatihan (latih[fitur] dan y).
# Menggunakan model RandomForestClassifier dari sklearn.ensemble.
klasifikasi.fit(latih[fitur], x)
```

[Dokumentasi Random Forest Classifier sklearn](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html#sklearn.ensemble.RandomForestClassifier)

```python
# Membuat variabel hasil_pengujian_probabilitas yang berisi nilai probabilitas dari hasil data pengujian.
hasil_pengujian_probabilitas = klasifikasi.predict_proba(uji[fitur])

# Membuat DataFrame baru untuk menampilkan hasil prediksi probabilitas pada data pengujian
hasil_pengujian_probabilitas_df = pd.DataFrame(hasil_pengujian_probabilitas, columns=['setosa(0)', 'versicolor(1)', 'virginica(2)'])

# Menampilkan DataFrame hasil prediksi probabilitas
print(hasil_pengujian_probabilitas_df)
```

```python
# Melakukan prediksi pada data pengujian dan di inisiasikan dalam variabel hasil_pengujian.
hasil_pengujian = klasifikasi.predict(uji[fitur])

# Menampilkan hasil prediksi pada data pengujian
# print(hasil_pengujian)

# Membuat DataFrame baru untuk menampilkan hasil prediksi pada data pengujian
hasil_pengujian_df = pd.DataFrame({'hasil_pengujian': hasil_pengujian})

# Menampilkan DataFrame hasil pengujian
print(hasil_pengujian_df)
```

```python
# Melakukan prediksi pada data pengujian dan menyimpan hasilnya dalam variabel preds.
# Menampilkan hasil prediksi dalam bentuk nama spesies menggunakan data_iris.target_names.
hasil_prediksi_nama_spesies = data_iris.target_names[klasifikasi.predict(uji[fitur])]

# Menampilkan hasil prediksi dari data pengujian
# print(preds)

# Membuat DataFrame baru untuk menampilkan hasil prediksi dalam bentuk nama spesies
hasil_prediksi_nama_spesies_df = pd.DataFrame({'hasil_prediksi_spesies': hasil_prediksi_nama_spesies})

# Menampilkan DataFrame hasil prediksi dalam bentuk nama spesies
print(hasil_prediksi_nama_spesies_df)
```

```python
# Menampilkan matriks kebingungan (confusion matrix) untuk mengevaluasi kinerja model.
# Memberikan gambaran tentang seberapa baik model melakukan klasifikasi.

pd.crosstab(uji['spesies'], hasil_prediksi_nama_spesies, rownames=['spesies asli'], colnames=['prediksi spesies'])
```

```python
# Menghitung dan mencetak presentase benar dan presentase salah dari model.
presentase_benar = (30 / 32) * 100
presentase_salah = (2 / 32) * 100

print(f'Presentase Benar Sebesar {presentase_benar}%')
print(f'Presentase Salah Sebesar {presentase_salah}%')
```

```python
# Terakhir, kode mencoba untuk memprediksi spesies berdasarkan beberapa sampel data yang diberikan dan mencetak hasil prediksi.
# 3,4.7,3.2,1.3,0.2,Iris-setosa
# 74,6.1,2.8,4.7,1.2,Iris-versicolor
# 118,7.7,3.8,6.7,2.2,Iris-virginica

new_data = np.array([[4.7, 3.2, 1.3, 0.2], [6.1, 2.8, 4.7, 1.2], [7.7, 3.8, 6.7, 2.2]])

# Membuat DataFrame baru untuk data baru dengan nama fitur yang sesuai
new_df = pd.DataFrame(new_data, columns=data_iris.feature_names)

# Melakukan prediksi pada data baru
predictions = klasifikasi.predict(new_df)

# Mengonversi hasil prediksi menjadi nama spesies
predicted_species = data_iris.target_names[predictions]

# Mencetak hasil prediksi
print(predicted_species)
```
