```python
# Memuat dataset Iris menggunakan load_iris() dari scikit-learn.
from sklearn.datasets import load_iris

# Mengimpor RandomForestClassifier dari sklearn.ensemble untuk menginisialisasi model Random Forest.
from sklearn.ensemble import RandomForestClassifier

# Mengimpor pandas untuk memanipulasi data dalam bentuk DataFrame.
import pandas as pd

# Mengimpor numpy dengan alias np.
import numpy as np

# Menetapkan seed acak dengan np.random.seed(0) untuk hasil yang dapat direproduksi.
np.random.seed(0)
```

```python
# Membuat objek iris yang berisi data Iris.
iris = load_iris()

# Membuat DataFrame df yang berisi data fitur Iris.
df = pd.DataFrame(iris.data, columns = iris.feature_names)

# Viewing the top 5 rows
df.head()
```

```python
# Menambahkan kolom 'species' ke DataFrame df untuk menampilkan nama spesies.
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Viewing the top 5 rows
df.head()
```

```python
# Membuat kolom 'is_train' yang berisi bilangan acak untuk membagi data menjadi data pelatihan dan data pengujian.
df['is_train'] = np.random.uniform(0, 1, len(df)) <= .75

# View the top 5 rows
df.head()
```

```python
# Membuat DataFrames terpisah untuk data pelatihan (train) dan data pengujian (test).
train, test = df[df['is_train']==True], df[df['is_train']==False]

# Menampilkan jumlah pengamatan dalam data pelatihan dan data pengujian.
print('Number of observations in the training data:', len(train))
print('Number of observations in the test data:', len(test))
```

```python
# Menentukan kolom fitur yang akan digunakan dalam pelatihan model.
features = df.columns[:4]

# View features
print(features)
```

```python
# Mengubah nama spesies menjadi angka menggunakan pd.factorize().
y = pd.factorize(train['species'])[0]

# viewing target
print(y)
```

```python
# Membuat objek RandomForestClassifier dengan parameter tertentu, seperti jumlah pekerjaan (n_jobs) dan seed acak.
clf = RandomForestClassifier(n_jobs=2, random_state=0)

# Melatih model dengan menggunakan data pelatihan.
clf.fit(train[features], y)
```

```python
# Melakukan prediksi pada data pengujian.

clf.predict(test[features])
```

```python
# Viewing the predicted probabilities of the first 10 observations
clf.predict_proba(test[features])[0:10]
```

```python
# Menampilkan hasil prediksi dalam bentuk nama spesies menggunakan iris.target_names.
preds = iris.target_names[clf.predict(test[features])]

# View the PREDICTED species for the first five observations
print(preds[:5])
```

```python
# Menampilkan matriks kebingungan (confusion matrix) untuk mengevaluasi kinerja model.

pd.crosstab(test['species'], preds, rownames=['actual species'], colnames=['Predicted Species'])
```

```python
# Terakhir, kode mencoba untuk memprediksi spesies berdasarkan beberapa sampel data yang diberikan dan mencetak hasil prediksi.
# 3,4.7,3.2,1.3,0.2,Iris-setosa
# 74,6.1,2.8,4.7,1.2,Iris-versicolor
# 118,7.7,3.8,6.7,2.2,Iris-virginica

print(iris.target_names[clf.predict( [[4.7, 3.2, 1.3, 0.2], [6.1, 2.8, 4.7, 1.2], [7.7, 3.8, 6.7, 2.2]] )])
```
